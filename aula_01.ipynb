### Equações que regem o movimento de um corpo submetido a um lançamento vertical.

Um corpo lançado verticalmente para cima, com uma determinada velocidade, é considerado como um movimento unidimensional, onde, por não afetar siginificativamente a trajetória, se desconsidera o efeito do atrito com o ar. Inicialmente, no movimento de subida do projétil, a ação da força devido ao campo gravitacional da terra, atua com uma acelaração no sentido aposto a direção seu deslocamento (movimento retardado). No momento em que o objeto atinge a altura máxima, o sentido da velocidade muda, o objeto inicia a queda livre (movimento retardado). Uma animação que representa esse movimento pode ser visualizada em ([ref](http://www.college-physics.com/book/mechanics/vertically-launched-projectiles/).)

As equações associadas a esse movimento são as seguintes. 


$$y(t) = y_{\text{o}} + v_{\text{oy}} t + \frac{a_{\text{y}}}{2} t^{2}  \ \ \ \ (1) $$

$$v_{\text{y}} = v_{\text{oy}} + a_{\text{y}} t  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (2)$$                       

$$ v_{\text{y}}^{2} = v_{\text{oy}}^{2} + 2 a_{\text{y}} \Delta y  \ \ \ \ \ \ \ \ \ \ \ \ (3)$$     

$$a_{\text{y}} = -g   \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (4)$$                                                  

Sendo $g = 9.806 \  \text{m/s}^{2}$ o valor da aceleração da gravidade ([ref](https://en.wikipedia.org/wiki/Gravitational_acceleration).)

Iremos nos concentrar na analise de uma tabela de dados de um determinado experimento de lançamento vertical.

##  Ajuste da função

A melhor maneira de encontrar o valor da aceleração de queda livre é * ajustar uma função * aos nossos dados. Esta é uma técnica experimental útil.

Suponha que tentamos encontrar a função polinomial

$$
p(t) = A t^2 + B t + C \approx y(t)
$$ 

que melhor corresponda ao nosso conjunto de dados. O Python numérico (`numpy`) pode fazer isso com a função `polyfit`. Você pode aprender os detalhes no [polyfit documentation](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.polyfit.html).

Esperamos que nossos dados de $ y $ versus $ t $ sigam um polinômio porque, em queda livre, a equação de movimento é

$$
y(t) = y_{\text{o}} + v_{\text{oy}} t - \frac{g}{2} t^{2}
$$

onde $ a $ é a aceleração devido à gravidade (também chamada de $ g $).

Vamos aplicar os nossos dados à função `polyfit` com os parâmetros A, B e C que ela ajuste a um polinômio de grau 2 - isto é, uma parábola. A função retorna os coeficientes $ A, B, C $ que melhor se ajustam aos nossos dados.

```python
A,B,C = polyfit(t,y,2)       # ajuste polinomial de grau 2 no dado t,y
print('Melhor ajuste com os coeficientes')
print('     A=',A,'B=',B,'C=',C)
```

```python
aceleração_gravidade = (2*abs(A)).round(3) #valor da gravidade com precisão de 3 casas
"aceleração gravitacional é {} m/s²".format(aceleração_gravidade)
```

Vamos ver o quão bem esse polinômio se ajusta aos nossos dados, os colocando juntos em um mesmo gráfico.

Para plotar a parábola, usaremos a função `linspace` para criar uma matriz de muitos valores de tempo, do índice` 0` até a última unidade determinada, `t [-1]`. Em seguida, conecte esses tempos no polinômio com os valores de `A, B, C` que encontramos acima.

```python
# exemplo do uso da função .format para o label

numer1=55
numer2=34
result12=numer1*numer2
valor_texto="o produto {} x {} é igual a {}".format(numer1,numer2,result12)
valor_texto
```

```python
value_in_string = "Ajuste da parabola $y(t)= {}t² + {}t - {}$".format(A.round(2), B.round(2), abs(C).round(2))
value_in_string # definindo label com a função format
```

```python
t_axis = linspace(t[0],t[-1],(f-i))     #  t_axis é um vetor de mesmo comprimento do dado
plt.grid()
pp = A*t_axis**2 + B*t_axis + C       # avaliar o polinômio para todos os valores de t_axis
plt.plot(t,y,'o',label='Dado')    # gráfico dos dados
plt.plot(t_axis,pp,label=value_in_string)  # gráfico da parábola para todos os valores de t_axis
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.legend(loc='lower center',shadow=True)        # Mostra o gráfico com a legenda value_in_string
plt.savefig('graf_medida_g')
```

### Qualidade do Ajuste

Iremos aplicar a nossa análise, o teste $R^{2}$, para termos, qualitativamente a informação sobre a relevancia estatística de nosso estudo. 

```python
import numpy as np

# Obtendo o valor de R2
yresid = y - pp # resíduo = valor real - valor ajustado (valor predito)
SQresid = sum(pow(yresid,2)) # soma dos quadrados dos resíduos 
SQtotal = len(y) * np.var(y) # número de elementos do vetor y vezes a variância de y
R2 = 1 - SQresid/SQtotal # coeficiente de determinação

print('Melhor ajuste com os coeficientes', 'A=', A,', B=',B,'e C=',C) 
```

```python
R2_texto="R² = {}".format(R2.round(5)) # definindo label com a função format 
R2_texto
```

### Derivada da função de ajuste

Derivando a função de ajuste $y(t) = - 4,\!9 t² + 5,\!18 t - 0,\!01 $, obtemos a função temporal
da velocidade

```python
from sympy import Symbol, Derivative

x= Symbol('x')  # define qual carcter será a variável

function= - 4.9*x**2 + 5.18*x + 0.01 # função horária a ser derivada

deriv= Derivative(function, x) # aplica a operação de derivada
deriv.doit() # mostra a função derivada da posição que é a função velocidade
```

## Aceleração gravitacional

Qual é a incerteza nos coeficientes de ajuste? Agora temos o coeficiente $ A $ desde o ajuste, mas queremos saber $ A \pm \delta A $, onde $ \delta A $ é a incerteza absoluta.

Quando a função `polyfit` é chamada com um parâmetro adicional:

    polyfit(t,y,2,cov=True)
    

retorna os coeficientes `A, B, C` como antes e também uma" matriz de covariância "que fornece a variação em cada um dos coeficientes de ajuste. A covariancia é o quadrado do desvio padrão e o desvio padrão é a incerteza no coeficiente de ajuste. A variância de cada coeficiente aparece na diagonal da matriz de covariância. Esses termos foram analisados nos módulos inicias desse curso.

Encontramos a incerteza em cada parâmetro de ajuste (`dA, dB, dC`) da seguinte forma:

```python
(A,B,C),covariance = polyfit(t,y,2,cov=True)        # Ajuste com a matriz de covariância
dA,dB,dC = sqrt(diag(covariance))
print('Os coeficientes de ajuste são')
print('   A = ',A,'±',dA,'m/s^2')
print('   B = ',B,'±',dB,'m/s')
print('   C = ',C,'±',dC,'m')
```

Agora você mostrar a medida da experiência para o valor da aceleração devido à queda livre, $g$, juntamente com a incerteza.


```python
g = abs(2*A)    # 
dg = 2*dA   # 
print('Medida do valor de g = ',g,'±',dg,'m/s²')

```
