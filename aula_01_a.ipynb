### Equações que regem o movimento de um corpo submetido a um lançamento vertical.

Um corpo lançado verticalmente para cima, com uma determinada velocidade, é considerado como um movimento unidimensional, onde, por não afetar siginificativamente a trajetória, se desconsidera o efeito do atrito com o ar. Inicialmente, no movimento de subida do projétil, a ação da força devido ao campo gravitacional da terra, atua com uma acelaração no sentido aposto a direção de seu deslocamento (movimento retardado). No momento em que o objeto atinge a altura máxima, o sentido da velocidade muda, o objeto inicia a queda livre (movimento retardado). Uma animação que representa esse movimento pode ser visualizada em ([ref](http://www.college-physics.com/book/mechanics/vertically-launched-projectiles/).)

As equações associadas a esse movimento são as seguintes. 


$$y(t) = y_{\text{o}} + v_{\text{oy}} t + \frac{a_{\text{y}}}{2} t^{2}  \ \ \ \ (1) $$

$$v_{\text{y}} = v_{\text{oy}} + a_{\text{y}} t  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (2)$$                       

$$ v_{\text{y}}^{2} = v_{\text{oy}}^{2} + 2 a_{\text{y}} \Delta y  \ \ \ \ \ \ \ \ \ \ \ \ (3)$$     

$$a_{\text{y}} = -g   \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (4)$$                                                  

Sendo $g = 9.806 \  \text{m/s}^{2}$ o valor da aceleração da gravidade ([ref](https://en.wikipedia.org/wiki/Gravitational_acceleration).)

Iremos nos concentrar na analise de uma tabela de dados de um determinado experimento de lançamento vertical.

##  Ajuste da função

A melhor maneira de encontrar o valor da aceleração de queda livre é * ajustar uma função * aos nossos dados. Esta é uma técnica experimental útil.

Suponha que tentamos encontrar a função polinomial

$$
p(t) = C + B t + A t^2 \approx y(t)
$$ 

que melhor corresponda ao nosso conjunto de dados. O Python numérico (`numpy`) pode fazer isso com a função `polyfit`. Você pode aprender os detalhes no [polyfit documentation](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.polyfit.html).

Esperamos que nossos dados de $ y $ versus $ t $ sigam um polinômio porque, em queda livre, a equação de movimento é

$$
y(t) = y_{\text{o}} + v_{\text{oy}} t - \frac{g}{2} t^{2}
$$

onde $ a $ é a aceleração devido à gravidade (também chamada de $ g $).

Vamos aplicar os nossos dados à função `polyfit` com os parâmetros A, B e C que ela ajuste a um polinômio de grau 2 - isto é, uma parábola. A função retorna os coeficientes $ A, B, C $ que melhor se ajustam aos nossos dados.

```python
A,B,C = polyfit(t,y,2)       # ajuste polinomial de grau 2 no dado t,y
print('Melhor ajuste com os coeficientes')
print('     A=',A,'B=',B,'C=',C)
```

```python
# exemplo do uso da função .format para o label

numer1=55
numer2=34
result12=numer1*numer2
valor_texto="o produto {} x {} é igual a {}".format(numer1,numer2,result12)
valor_texto
```

```python
aceleração_gravidade = (2*abs(A)).round(3) #valor da gravidade com precisão de 3 casas
"aceleração gravitacional é {} m/s²".format(aceleração_gravidade)
```

Vamos ver o quão bem esse polinômio se ajusta aos nossos dados, os colocando juntos em um mesmo gráfico.

Para plotar a parábola, usaremos a função `linspace` para criar uma matriz de muitos valores de tempo, do índice` 0` até a última unidade determinada, `t [-1]`. Em seguida, conecte esses tempos no polinômio com os valores de `A, B, C` que encontramos acima.

```python
value_in_string = "Ajuste da parabola $y(t)= {}t² + {}t - {}$".format(A.round(2), B.round(2), abs(C).round(2))
value_in_string # definindo label com a função format
```

```python
t_axis = linspace(t[0],t[-1],(f-i))     #  t_axis é um vetor de mesmo comprimento do dado
plt.grid()
pp = A*t_axis**2 + B*t_axis + C       # avaliar o polinômio para todos os valores de t_axis
plt.plot(t,y,'o',label='Dado')    # gráfico dos dados
plt.plot(t_axis,pp,label=value_in_string)  # gráfico da parábola para todos os valores de t_axis
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.legend(loc='lower center',shadow=True)        # Mostra o gráfico com a legenda value_in_string
plt.savefig('graf_medida_g')
plt.title('Fig. 3:  Altura vs Tempo e função ajuste.',y=-0.2);
```

###  Ajuste dos dados: $R^{2}$

Iremos aplicar a nossa análise, o teste $R^{2}$, para termos, qualitativamente a informação sobre a relevancia estatística de nosso estudo. 

```python
# Obtendo o valor de R2   

s_res = sum((y - pp)**2) # resíduo = (valor real - valor ajustado (valor predito))   
s_tot = sum((y - y.mean())**2)       # calcula a soma total
R2 = 1 - s_res/s_tot # retorna o coeficiente de determinação
R2
```

```python
R2_texto="R² = {}".format(R2.round(5)) # definindo label com a função format 
R2_texto
```

```python
plt.grid()
plt.plot(t,y,'o',label='dado')    # # gráfico dos dados
plt.plot(t_axis,pp,label=value_in_string)  # gráfico da parábola para todos os valores de t_axis
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.legend(loc="lower center", ncol=1, shadow=True, title=R2_texto)
```

```python
# ajustando algumas configurações para melhor visualização dos gráficos
plt.rcParams['font.size'] = 16                           # altera o tamanho das fontes na figura
plt.rcParams['figure.figsize'] = 10, 8                    # altera o tamanho da figura
plt.rcParams['axes.axisbelow'] = True                    # para desenhar os grids atrás das curvas e pontos
for axis in ['xtick', 'ytick']:                          # loop para aplicar as mudanças nos eixos x e y
    plt.rcParams['{}.major.size'.format(axis)] = 10      # altera o tamanho dos ticks maiores
    plt.rcParams['{}.minor.size'.format(axis)] = 5       # altera o tamanho dos ticks menores
    plt.rcParams['{}.minor.visible'.format(axis)] = True # faz os ticks menores ficarem visíveis

t_all = array([0.    , 0.05  , 0.11  , 0.16  , 0.2   , 0.25  , 0.31  , 0.351 ,
       0.401 , 0.45  , 0.5   , 0.551 , 0.601 , 0.65  , 0.7   , 0.75  ,
       0.801 , 0.8501, 0.9   , 0.95  , 1.    , 1.0501])

y_all= array([0.08  , 0.1501, 0.498 , 0.683 , 0.829 , 0.978 , 1.1182, 1.203 ,
       1.278 , 1.329 , 1.355 , 1.357 , 1.334 , 1.288 , 1.217 , 1.118 ,
       0.998 , 0.852 , 0.682 , 0.488 , 0.268 , 0.021])


def z(i,f):
    t = t_all[i:f]                # t é o corte para a coluna t_all
    y = y_all[i:f]                # y é o corte para a coluna y_all
    A,B,C = polyfit(t,y,2)       # ajuste polinomial de grau 2 no dado t,y
    t_axis = linspace(t[0],t[-1],(f-i))     #  t_axis é um vetor de mesmo comprimento do dado
    value_in_string = "Ajuste da parabola $y(t)= {}t² + {}t - {}$".format(A.round(2), B.round(2), abs(C).round(2)) #str
    pp = A*t_axis**2 + B*t_axis + C       # avaliar o polinômio de ajuste para todos os valores de t_axis
    s_res = sum((y - pp)**2) # resíduo = (valor real - valor ajustado (valor predito))   
    s_tot = sum((y - y.mean())**2)       # calcula a soma total
    R2 = 1 - s_res/s_tot # retorna o coeficiente de determinação
    R2_texto="R² = {}".format(R2.round(5)) # definindo label com a função format     
    plt.plot(t,y,'o',label='Dado')    # gráfico dos dados
    plt.plot(t_axis,pp,label=value_in_string)  # gráfico da parábola para todos os valores de t_axis
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição, y (m)')
    plt.legend(loc="lower center", ncol=1, shadow=True, title=R2_texto)        # posição da legenda //right
    plt.grid()
    plt.show()
    
interactive_plot = interactive(z, i=(0, 5, 1), f=(18, 22, 1))
output = interactive_plot.children[-1]
interactive_plot 
```

## Velocidade 

### Derivada da função de ajuste

Derivando a função de ajuste $y(t) $, obtemos a função temporal
da velocidade

```python
from sympy import Symbol, Derivative

x= Symbol('x')  # define qual carcter será a variável

function= - 4.88*x**2 + 5.16*x - 0.01 # função horária a ser derivada

deriv= Derivative(function, x) # aplica a operação de derivada
deriv.doit() # mostra a função derivada da posição que é a função velocidade
```

$$v_{\text{y}} = a_{\text{y}}  t + v_{\text{oy}} $$

```python
#ajustando algumas configurações para melhor visualização dos gráficos
plt.rcParams['font.size'] = 16                           # altera o tamanho das fontes na figura
plt.rcParams['figure.figsize'] = 22, 8                    # altera o tamanho da figura
plt.rcParams['axes.axisbelow'] = True                    # para desenhar os grids atrás das curvas e pontos
for axis in ['xtick', 'ytick']:                          # loop para aplicar as mudanças nos eixos x e y
    plt.rcParams['{}.major.size'.format(axis)] = 10      # altera o tamanho dos ticks maiores
    plt.rcParams['{}.minor.size'.format(axis)] = 5       # altera o tamanho dos ticks menores
    plt.rcParams['{}.minor.visible'.format(axis)] = True # faz os ticks menores ficarem visíveis

t_all = array([0.    , 0.05  , 0.11  , 0.16  , 0.2   , 0.25  , 0.31  , 0.351 ,
       0.401 , 0.45  , 0.5   , 0.551 , 0.601 , 0.65  , 0.7   , 0.75  ,
       0.801 , 0.8501, 0.9   , 0.95  , 1.    , 1.0501])

y_all= array([0.08  , 0.1501, 0.498 , 0.683 , 0.829 , 0.978 , 1.1182, 1.203 ,
       1.278 , 1.329 , 1.355 , 1.357 , 1.334 , 1.288 , 1.217 , 1.118 ,
       0.998 , 0.852 , 0.682 , 0.488 , 0.268 , 0.021])


def z(i,f):
    t = t_all[i:f]                # t é o corte para a coluna t_all
    y = y_all[i:f]                # y é o corte para a coluna y_all
    A,B,C = polyfit(t,y,2)       # ajuste polinomial de grau 2 no dado t,y
    t_axis = linspace(t[0],t[-1],(f-i))     #  t_axis é um vetor de mesmo comprimento do dado
    value_in_string = "Ajuste da parabola $y(t)= {}t² + {}t - {}$".format(A.round(2), B.round(2), abs(C).round(2)) #str
    pp = A*t_axis**2 + B*t_axis + C       # avaliar o polinômio de ajuste para todos os valores de t_axis
    s_res = sum((y - pp)**2) # resíduo = (valor real - valor ajustado (valor predito))   
    s_tot = sum((y - y.mean())**2)       # calcula a soma total
    R2 = 1 - s_res/s_tot # retorna o coeficiente de determinação
    R2_texto="R² = {}".format(R2.round(5)) # definindo label com a função format     
    plt.subplot(121)
    plt.plot(t,y,'o',label='Dado')    # gráfico dos dados
    plt.plot(t_axis,pp,label=value_in_string)  # gráfico da parábola para todos os valores de t_axis
    plt.xlabel('Tempo (s)')
    plt.ylabel('Posição, y (m)')
    plt.legend(loc="lower center", ncol=1, shadow=True, title=R2_texto)        # posição da legenda //right
    plt.grid()
    plt.subplot(122)
    vt = 2*A*t_axis + B       # avalia a função linear em cada valor de t_vaxis
    vel_texto = "função velocidade v(t) = {}t + {}".format((2*A).round(2),B.round(2))# definindo label com a função format
    plt.plot(t_axis,vt,label=vel_texto)  # gráfico da função horária da velocidade
    plt.xlabel('Tempo (s)')
    plt.ylabel('Velocidade (m/s)')
    plt.legend(loc='lower center',shadow=True)        # posição da legenda
    plt.grid()
    plt.show()
    
interactive_plot = interactive(z, i=(0, 7, 1), f=(18, 23, 1))
#output = interactive_plot.children[-1]
interactive_plot 
```
